```python
from scrapers.LBA import LBAScraper
import utils
parser = LBAScraper()
parser.current_game = {'game_id': 24093}

url = f'https://www.legabasket.it/game/{parser.current_game["game_id"]}'
soup = utils.get_soup(url)
parser.starters = parser.get_starters(soup)
print(parser.starters['home'])
```

```python
boxes = parser.get_boxes(soup)
```

```python
import pandas as pd

players_df = pd.DataFrame(
    columns=['Team', 'Player', 'MIN', 'PTS', 'P2M', 'P2A', 'P3M', 'P3A', 'FTM', 'FTA', 'OREB', 'DREB',
                'AST', 'TOV', 'STL', 'BLK', 'PF', 'PM'])
team_df = pd.DataFrame(
    columns=['Team', 'MIN', 'PTS', 'P2M', 'P2A', 'P3M', 'P3A', 'FTM', 'FTA', 'OREB', 'DREB', 'AST', 'TOV',
                'STL', 'BLK', 'PF', 'PM'])
opponent_df = pd.DataFrame(
    columns=['Team', 'MIN', 'PTS', 'P2M', 'P2A', 'P3M', 'P3A', 'FTM', 'FTA', 'OREB', 'DREB', 'AST', 'TOV',
                'STL', 'BLK', 'PF', 'PM'])
for team in boxes:
    players_df = pd.concat([players_df, pd.DataFrame(boxes[team]['players'])], ignore_index=True)
    team_df = pd.concat([team_df, pd.DataFrame(boxes[team]['team'])], ignore_index=True)
    opponent_df = pd.concat([opponent_df, pd.DataFrame(boxes[team]['opponent'])], ignore_index=True)
```

```python
ingressi_home = 0
uscite_home = 0
ingressi_away = 0
uscita_away = 0

for action in actions:
    if action['description'] == 'Ingresso':
        if action['home_club'] == 0:
            ingressi_home += 1
        else:
            ingressi_away += 1
        print(f"Club: {action['home_club']} - {action['description']}: period: {action['period']} {action['minute']}:{action['seconds']}")
    if action['description'] == 'Uscita':
        if action['home_club'] == 0:
            uscite_home += 1
        else:
            uscita_away += 1
        print(f"Club: {action['home_club']} - {action['description']}: period: {action['period']} {action['minute']}:{action['seconds']}")
```

```python
print(ingressi_home, uscite_home)
print(ingressi_away, uscita_away)
```
